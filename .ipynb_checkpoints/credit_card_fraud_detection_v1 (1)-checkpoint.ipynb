```python
492 out of 284, 807 rows are fraud data, which is 0.173%; 284, 315 out of 284, 807 rows are non fraud data, which is 99.827%.
This shows that this dataset is unbalanced dataset. 
```

```python
#visulize the unbalanced data
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot('Class', data=df)
plt.title('Class Distributions \n (0 = non-fraud and 1 = fraud)', fontsize=14)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Make sure we use the subsample in our correlation

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(24,20))

# Entire DataFrame
corr = df_1.corr()
sns.heatmap(corr, cmap='coolwarm_r', annot_kws={'size':20}, ax=ax1)
ax1.set_title("Fraud data Correlation Matrix \n", fontsize=14)


sub_sample_corr = df_0.corr()
sns.heatmap(sub_sample_corr, cmap='coolwarm_r', annot_kws={'size':20}, ax=ax2)
ax2.set_title('Non fraud data Correlation Matrix \n', fontsize=14)
plt.show()
```
